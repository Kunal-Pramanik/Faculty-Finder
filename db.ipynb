```python
import pandas as pd
from sqlalchemy import create_engine,text
```

```python
username = "postgres"
password = "0987"
host = "localhost"
port = "5432"
database = "Faculty_finder"
# Create SQLAlchemy engine
engine = create_engine(f'postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}')
print("connected")
```

```python
df = pd.read_csv('D:\BDE\Project\dau_faculty_clean.csv')
```

```python
faculty_df = df.rename(columns={
    "Faculty_id": "faculty_id",
    "Name": "name",
    "Education": "education",
    "Research_Interests": "research_interests",
    "Profile_URL": "profile_url",
    "clean_text": "clean_text"
})
```

```python
faculty_df.to_sql(
    "faculty",
    con=engine,
    schema="faculty_list",
    if_exists="append",
    index=False
)
```

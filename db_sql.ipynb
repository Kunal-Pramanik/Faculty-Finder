```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("faculty.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS faculty (
    faculty_id TEXT PRIMARY KEY,
    name TEXT,
    profile_url TEXT,
    education TEXT,
    email TEXT,
    contact_number TEXT,
    research_area TEXT
)
""")
```

```python
dau_faculty = pd.read_csv("dau_faculty.csv")
```

```python
dau_faculty.to_sql(
    "faculty",
    conn,
    if_exists="append",
    index=False
)
```
